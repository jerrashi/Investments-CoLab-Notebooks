Part B:
False - CAPM uses 2 factors, mean and variance. However, the CAPM final solution does use just one factor:
\begin{align}
̃̃r_{i,t} = r_f + β_i(̃r_{m,t} - r_f) + ε_{i,t}\\
\end{align}
Thus, the conclusion leads to a 1 factor model; however, the assumptions require two factors, variance and mean.

True - In APT, alpha of a stock represents expected returns. An asset should not have negative expected returns so the APT does not hold.

Part D:
False - In this case, the R^2 is measuring how much of the variation in expected return can be explained by variations in the different risk factors of the APT model.
