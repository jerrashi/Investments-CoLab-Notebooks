\begin{align}
var(\tilde{r_p})&=w^2_1σ^2_1 + w^2_2σ^2_2 + 2w_1w_2σ_{1,2}\\

\begin{align}
E[\tilde{r_p}]&=w_AE(\tilde{r_A})+w_BE(\tilde{r_B})\\

**Part C)** To calculate the Sharpe ratio of a portfolio, we use the following formula (also from slide 25 in lecture 4):

\begin{align}
\frac{dE[̃r_p]}{dSD[̃r_p]} &= \frac{\frac{dE[̃r_p]}{dw}}{\frac{dSD[̃r_p]}{dw}}\\
&= \frac{dE[̃r_{mkt}]-r_f}{σ_{mkt}}\\
&= \frac{Return - Risk-Free Return}{Standard Deviation}\\
\end{align}

We know that the Expected Return of each possible portfolio mix of B & C is the y axis and the Standard Deviation is the x axis. Thus we can add a Sharpe Ratio curve to our existing graph.

\begin{align}
y &= mx + b\\
19 &= \frac{19-10}{10}25+b\\
b &= -3.5\\
y &= 0.9x - 3.5\\
\end{align}

````python
# Same graph as before
# Added a Sharpe Ratio curve using the equation we just solved
# Used the following tutorial
# https://www.geeksforgeeks.org/plot-mathematical-expressions-in-python-using-matplotlib/

# importing the required module
import matplotlib.pyplot as plt
import numpy as np
 
# x axis values (standard deviation)
x1 = [0,15,25]
# corresponding y axis values (expected return)
y1 = [4,13,19]
 
# plotting the points
plt.plot(x1, y1, label = "Capital Allocation Line")
plt.scatter([0, 15, 25], [4, 13, 19])
plt.scatter([15], [10])

#NEW - sharpe ratio line
x2 = np.linspace(15, 25, 11)
y2 = np.linspace(10, 19, 11)
sharpe = ((0.9 * x2 - 3.5)-4)/x2
plt.plot(x2, y2, label = "B & C Opportunity Set")
plt.plot(x2, sharpe, label = "Sharpe Ratio")
plt.scatter([15, 25], [0.4, 0.6])


# naming the x axis
plt.xlabel('S (standard dev)')
# naming the y axis
plt.ylabel('E (expected return')
#labeling data points, specifying distance from point & alignment
plt.annotate("Portfolio A", (0, 4), textcoords="offset points", xytext=(0,10), 
             ha='left')
plt.annotate("Mixed Portfolio", (15, 13), textcoords="offset points", 
             xytext=(0,10), ha='center')
plt.annotate("Portfolio B", (25, 19), textcoords="offset points", 
             xytext=(0,-15), ha='right')
plt.annotate("Portfolio C", (15, 10), textcoords="offset points", 
             xytext=(0,10), ha='center')
#NEW - label Sharpe co-ordinates
#bit of a janky way to label co-ordinates but I didn't have time to figure out
#proper syntax
plt.annotate("(15, 0.4)", (15, 0.4), textcoords="offset points", 
             xytext=(0,10), ha='center')
plt.annotate("(25, 0.6)", (25, 0.6), textcoords="offset points", 
             xytext=(0,10), ha='center')
 
# giving a title to my graph
plt.title('Investment Opportunity Set')


# show a legend on the plot
plt.legend()
 
# function to show the plot
plt.show()
````
